```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
```

```python
gdp = pd.ExcelFile("indicator gapminder gdp_per_capita_ppp.xlsx")
```

```python
output_notebook()
```

```python
gdp = gdp.parse("Data")
```

```python
gdp=gdp.rename(columns = {'GDP per capita':'country'})
gdp = gdp[(gdp["country"]=="India") | (gdp["country"]=="China") | (gdp["country"]=="Russia")]
gdp = gdp.set_index('country')
```

```python
years = gdp.columns.tolist()
```

```python
colors = 216*["red","orange","green"]
p = figure(plot_width=800, plot_height=800)
p.circle(gdp.loc["India"], years,color='red',size=5,legend="India")
p.circle(gdp.loc["China"], years,color='yellow',size=5,legend="China")
p.circle(gdp.loc["Russia"], years,color='green',size=5,legend="Russia")
p.add_tools(HoverTool(show_arrow=False, line_policy='next', tooltips=[
    ('gdp', '@x'),
    ('year', '@y')
]))
p.xaxis.axis_label = "GDP per capita"
p.yaxis.axis_label = "Year"
#r.hover_glyph = 
show(p)
```
